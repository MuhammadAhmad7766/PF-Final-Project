### _**Data Cleaning**_

```python
# it can also be used for rows
df. rename(columns={"iPhone Sales (in million units)":"iPhone Sales","iPad Sales (in million units)":"iPad Sales","Mac Sales (in million units)":"Mac Sales","Wearables (in million units)":"Wearables units"}, inplace=True)
df
```

```python
df.dropna()
```

```python
df.fillna(10)
```

```python
df['In.mean']=df['iPhone Sales'].fillna(df['iPhone Sales'].mean())
df['In.mean']
df
```

```python
df.drop(['In.mean'],axis=1,inplace=True)
df
```

```python
# To drop columns
del df['Wearables units']
df
```

```python
# To drop rows
df = df.drop(4)
df
```
